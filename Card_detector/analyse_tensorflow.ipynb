# Analyze of the tensorflow training on cards dataset

```python
import matplotlib.pyplot as plt

import os
import numpy as np
import pandas as pd

import cv2
from PIL import Image
from IPython.display import display

%matplotlib inline
```

```python
DIR_PATH = os.getcwd()
DATA_ANALYZE_PATH = 'resources/tensorflow_loss_data.csv'
DATA_ANALYZE_EVAL_PATH = 'resources/tensorflow_eval_loss.csv'
IMG_PATH = 'resources/img/'
```

```python
loss_data = pd.read_csv(DATA_ANALYZE_PATH)
del loss_data['Wall time']
```

```python
print(loss_data.columns)
loss_data.head(15)
```

```python
loss_data_count = loss_data.count()
print(loss_data_count)
```

```python
tr = loss_data.plot(x='Step', y='Value', figsize=(10,5), grid=True)
```

## Eval loss comparison

```python
loss_eval = pd.read_csv(DATA_ANALYZE_EVAL_PATH)
del loss_eval['Wall time']
```

```python
print(loss_eval.columns)
loss_eval.head(15)
```

```python
loss_eval_count = loss_eval.count()
print(loss_eval_count)
```

```python
ev = loss_eval.plot(x='Step', y='Value',figsize=(10,5), grid=True, color="red")
```
